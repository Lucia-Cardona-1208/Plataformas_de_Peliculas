#Importar librer√≠a y lectura de archivo

```python
#importaci√≥n de Pandas y lectura de archivo
import pandas as pd
csv_path=('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ-ts_lsvOSHcT_XBvoev_kIyV1cotw044BeDn3oehs7WLujetsLsqP9CFBTr_sfrYSWgejyni0S9Rg/pub?gid=299857629&single=true&output=csv')
df = pd.read_csv(csv_path, sep=",")

# Imprimir las primeras cinco filas de un dataframe para probar que todo est√° bien
df.head()
```

#Tipos de variables

```python
#Ver los tipos de valores de cada columna
df.dtypes
```

```python
df['Netflix'] = df['Netflix'].astype('string')
df['Hulu'] = df['Hulu'].astype('string')
df['Prime Video'] = df['Prime Video'].astype('string')
df['Disney+'] = df['Disney+'].astype('string')
df.dtypes
```

#Reemplazar datos en las columnas

```python
#Remplazar "0" por "no" y "1" por "si" en las columnas "Netflix" "Hulu" "Prime Video" "Disney+"
df['Netflix'] = df['Netflix'].replace({"0":"No", "1":"Si"})
df['Hulu'] = df['Hulu'].replace({"0":"No", "1":"Si"})
df['Prime Video'] = df['Prime Video'].replace({"0":"No", "1":"Si"})
df['Disney+'] = df['Disney+'].replace({"0":"No", "1":"Si"})
```

#Cambiar t√≠tulos

```python
#Cambiar t√≠tulos a las columnas
df.columns= ['Enumeraci√≥n','ID','T√≠tulo_Pel√≠cula', 'A√±o', 'Edad', 'Calificaci√≥n(tomat√≥metro)', 'Netflix', 'Hulu', 'Prime_Video', 'Disney+', 'Type']
```

```python
# Imprimir las primeras cinco filas de un dataframe para probar que todo est√° bien
df.head()
```

#Eliminar columnas

```python
#Eliminar columnas
df=df.drop('Enumeraci√≥n',axis=1)
df=df.drop('ID',axis=1)
df=df.drop('Type',axis=1)
```

```python
#Nos ayuda para ver los nombres de las columnas
df.columns
```

```python
# Imprimir las primeras cinco filas de un dataframe para probar que todo est√° bien
df.head()
```

```python
#Con este eliminamos las filas con datos nulos
df=df.dropna(axis=0)
```

```python
# Imprimir las primeras cinco filas de un dataframe para probar que todo est√° bien
df.head()
```

#Histograma y diagrama

```python
# Importando las bibliotecas necesarias para visualizaci√≥n de datos y an√°lisis exploratorio
import matplotlib.pyplot as plt  # Biblioteca para crear gr√°ficos y visualizaciones est√°ticas, animadas e interactivas en Python
import seaborn as sns  # Biblioteca para crear gr√°ficos estad√≠sticos atractivos e informativos en Python
```

```python
# Utilizando Seaborn para crear un diagrama de distribuci√≥n (histograma y KDE) para la columna 'A√±o' del DataFrame df
sns.distplot(df['A√±o'])
```

```python
# Creando un histograma para la columna 'A√±o' del DataFrame df
# figsize=(6,6) especifica las dimensiones del gr√°fico en pulgadas
df['A√±o'].hist(figsize=(6,6))

# Mostrando el gr√°fico en pantalla
plt.show()
```

```python
# Creando una figura con dimensiones espec√≠ficas (10x10 pulgadas)
fig=plt.figure(figsize=(20,50))

# Creando un diagrama de caja (boxplot) para comparar la distribuci√≥n de los precios entre los diferentes estilos de carrocer√≠a
sns.boxplot(x="Edad", y="A√±o", data=df)

# Mostrando el gr√°fico en pantalla
plt.show()
```
